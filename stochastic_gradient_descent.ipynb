```python
import numpy as np
```

```python
def score(y,y_pred):
    u=((y-y_pred)**2)
```

```python
def cost(points,m):
    N=points.shape[1]-1
    M=points.shape[0]
    x=points[:,0:N]
    y=points[:,N].reshape(-1,1)
    return ((y-(x@m))**2).mean()
```

```python
def step_gd(points,learning_rate,m):
    N=points.shape[1]-1
    M=points.shape[0]
    
    for i in range(M):
        y=points[i,N]
        x=points[i,0:N]
        slope=np.zeros((N,1))
        for j in range(N):
            slope[j] = (-2/M)*(y-(x@m).sum())*(x[j])
        m = m - learning_rate*slope
    return m
```

```python
def batch_gd(points,learning_rate,num_iterations):
    N=points.shape[1]-1
    M=points.shape[0]
    m=np.zeros((N,1))
    for i in range(num_iterations):
        m = step_gd(points,learning_rate,m)
        print(i," Cost ",cost(points,m))
    return m
```

```python
def run():
    data=np.loadtxt("data.csv",delimiter=",")
    learning_rate=0.01
    num_iterations=1000
    m = batch_gd(data,learning_rate,num_iterations)
    return m
```

```python
def create_data(m1,m2,m3):
    n=100
    mat=np.zeros((n,5))
    for i in range(n):
        mat[i,0]=np.random.randint(10)
        mat[i,1]=np.random.randint(10)
        mat[i,2]=np.random.randint(10)
        mat[i,4]=m1*mat[i,0]+m2*mat[i,1]+m3*mat[i,2]+np.random.randn()
        mat[i,3]=1
    np.savetxt("data.csv",mat,delimiter=",")
        
```

```python
create_data(-1,2,3)
```

```python
points=np.loadtxt("data.csv",delimiter=",")
x=points[:,0:4]
y=points[:,4].reshape(-1,1)
m=run()
y_pred=x@m
```

```python
import matplotlib.pyplot as plt
plt.scatter(y,y_pred)
plt.axis([0,100,1,100])
plt.show()
```
